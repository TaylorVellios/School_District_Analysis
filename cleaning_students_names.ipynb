```python
import pandas as pd

file_open = 'Resources\students_complete.csv'
students_df = pd.read_csv(file_open)
students_df
```

```python
# TAYLOR METHOD - WITHOUT PANDAS

shitty_names = [i for i in students_df["student_name"] if len(i.split()) > 2]
shitty_names
```

```python
#----TO GET THE STUDENT NAMES INTO A LIST USING PANDAS--- USE .tolist() on the "student_names" column
#Module Style To Filter Names
names = students_df["student_name"].tolist()

names_to_fix = []
for i in names:
    if len(i.split()) > 2:
        names_to_fix.append(i)
        
names_to_fix
```

```python
# CREATE A LIST OF ALL PREFIXES THAT ARE 4 OR LESS CHARACTERS IN LENGTH

#TAYLOR MODE
# prefixes = [i.split()[0] for i in names_to_fix if len(i.split()[0]) <= 4]

prefix = []
for name in names_to_fix:
    if len(name.split()[0]) <= 4:
        prefix.append(name.split()[0])
prefix
```

```python
# GET THE SUFFIXES NOW

#TAYLOR MODE
#suffixes = [i.split()[-1] for i in names_to_fix if len(i.split()[-1]) <= 3]

suffix = []
for name in names_to_fix:
    if len(name.split()[-1]) <= 3:
        suffix.append(name.split()[-1])
suffix
```

```python
#USE SET TO CRUSH THESE LONGLISTS

set(prefix)
```

```python
set(suffix)
```

```python
prefixes_suffixes = ['Ms. ','Miss ', 'Mrs. ', 'Mr. ', 'Dr. ', ' DDS', ' DVM', ' MD', ' PhD']

#.STRIP() vs .REPLACE()
#.strip() will remove any of the characters put in its argument, if we put "Mrs.", the output will remove "M"s "r"s and "S"s
#.replace takes 2 arguments, the string to search for, and then the string to replace it with.
    #--since we are removing prefixes/suffixes that are followed/proceeded by whitespace, 
    # all .remove arguments need a whitespace to go
for i in prefixes_suffixes:
    students_df['student_name'] = students_df['student_name'].str.replace(i, '')

students_df
```

```python
students_fixed = []
student_names = students_df['student_name'].tolist()

for name in student_names:
    if len(name.split()) >= 3:
        students_fixed.append(name)
students_fixed
```
